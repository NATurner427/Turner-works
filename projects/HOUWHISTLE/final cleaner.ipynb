```python
path='C://Users/16142/Desktop/Astro/2018 season/'
df=pd.read_csv(f'{path}2018 season pitch log.csv')
summary=pd.read_csv(f'{path}HOU_18_games.csv')
```

```python
df.head()
```

```python
date={'Mar':'03','Apr':'04','May':'05','Jun':'06','Jul':'07','Aug':'08','Sep':'09','Oct':'10'}
```

```python
lst=[]
for i,row in df.iterrows():
    split=row[1].split('-')
    lst.append(f'20{split[2]}-{date[split[1]]}-{split[0].zfill(2)}')
```

```python
df['Date']=lst
```

```python
pitches={1:['Two-Seam Fastball','Four-Seam Fastball','Cutter','Sinker','Splitter','Forkball'],
       2:['Changeup','Curveball','Slider','Knuckle Curve','Eephus','Screwball','Knuckleball'],
        3:'Pitchout'}

lst=[]
for i,row in df.iterrows():
    for sign,pitch in pitches.items():
            if row[8] in pitch:
                lst.append(sign)
                
df.insert(loc=9,column='pitch_category',value=lst)
```

```python
lst=[]
for i,row in df.iterrows():
    lst.append(str(datetime.timedelta(seconds=row[21])))

df.insert(loc=23,column='youtube_timestamp',value=lst)
```
