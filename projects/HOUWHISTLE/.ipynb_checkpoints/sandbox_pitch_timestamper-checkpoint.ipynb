I have all the videos of the games but I need to clip out the audio around the times of each pitch for each game. I could watch each game, and take note of when each pitch was thrown and then clip around that time, but typically 300 pitches are thrown each game, and there are 162regular season games plus 8 post season games to go through. 170*300= 51,000 pitches.

So, I will be using openCV to watch the games for me. I need something consistent in each game that changes when a pitch is thrown. At first I considered having the AI focus on the motion of the pitcher to register a pitch, and I believe this possible, but I have an easier option. I'll be focusing on the pitch count on the scorebox. this ticks up each time a pitch is thrown. this has the added benefit of also allowing me to match up the pitch number with the outcome of the pitch that pulled from baseball savant.

```python
%%time

try:
    from PIL import Image
except ImportError:
    import Image

import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
import datetime
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

got=0
lst=[]
consecutive=0

n=0
skip=20

#IMPORT TEMPLATES AND GRAB DIMENSIONS
box = cv2.imread('E:/HOU18/game 2/snapshots/full box.jpg',0)  
height, width = box.shape

p=cv2.imread('E:/HOU18/game 2/snapshots/pitch p.jpg',0) 
height_p, width_p = p.shape

#IMPORT VIDEO
cap = cv2.VideoCapture('E://HOU18/test videos/test video.mp4')


if cap.isOpened()== False: 
    print("Error opening the video file. Please double check your file path for typos")

while cap.isOpened():
    
    cap.set(cv2.CAP_PROP_POS_FRAMES, n)
    # Read a new frame
    ret, frame = cap.read()
    
    if ret == True:
        
        grayFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        tessFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # FIND INFOGRAPHIC
        res = cv2.matchTemplate(grayFrame,box,cv2.TM_CCOEFF_NORMED)
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
        
        top_left = max_loc
        bottom_right = (top_left[0] + width, top_left[1] + height)
        
        #FIND PITCH COUNT P
        res_p =cv2.matchTemplate(grayFrame,p,cv2.TM_CCOEFF_NORMED)
        min_val_p, max_val_p, min_loc_p, max_loc_p = cv2.minMaxLoc(res_p)
        
        top_left_p = max_loc_p
        bottom_right_p = (top_left_p[0] + width_p, top_left_p[1] + height_p) 
        
        #See if infographic is on screen
        if top_left[0] >= 1013 and top_left[0] <=1015 and top_left[1] >= 497 and top_left[1] <=499:
            cv2.rectangle(frame,top_left, bottom_right,(0,200,200), 3)
            
            #see if p: is in correct location
            if top_left_p[1] == 499:
                consecutive+=1
                
                # if consecutive frames have the p in correct location take time
                if consecutive==7:
                    cv2.rectangle(frame,top_left,(top_left_p[0],bottom_right_p[1]),(255,80,80), 2)
                    cv2.rectangle(frame,(bottom_right_p[0],top_left[1]),(bottom_right[0],bottom_right_p[1]),(80,80,255), 2)
                    cv2.rectangle(frame,top_left_p, bottom_right_p,(0,255,0), 1)
                    cv2.putText(frame, "pitch recorded", (1014,478), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0),2)
                    
                    cropFramePitcher= tessFrame[top_left[1]:bottom_right_p[1],top_left[0]:top_left_p[0]]
                    cropFramePitcher=cv2.cvtColor(cropFramePitcher,cv2.COLOR_BGR2GRAY)
                    
                    scale=1200
                    w = int(cropFramePitcher.shape[1] * scale / 100)
                    h = int(cropFramePitcher.shape[0] * scale/ 100)
                    dim = (w, h)
                    cropFramePitcher=cv2.resize(cropFramePitcher,dim,cv2.INTER_AREA)
                    cropFramePitcher=cv2.medianBlur(cropFramePitcher,9)
                    
                    r,cropFramePitcher=cv2.threshold(cropFramePitcher, 127, 255, cv2.THRESH_BINARY_INV)
                    
                    
                    
                    cropFrameCount= tessFrame[top_left[1]:bottom_right_p[1],bottom_right_p[0]-2:bottom_right[0]]
                    cropFrameCount=cv2.cvtColor(cropFrameCount,cv2.COLOR_BGR2GRAY)
                    
                    w = int(cropFrameCount.shape[1] * scale / 100)
                    h = int(cropFrameCount.shape[0] * scale/ 100)
                    dim = (w, h)
                    cropFrameCount=cv2.resize(cropFrameCount,dim,cv2.INTER_AREA)
                    cropFrameCount=cv2.medianBlur(cropFrameCount,19)
                    
                    r,cropFrameCount=cv2.threshold(cropFrameCount, 127, 255, cv2.THRESH_BINARY_INV)
                    
                    pitcher=pytesseract.image_to_string(cropFramePitcher)
                    count=pytesseract.image_to_string(cropFrameCount,config='--psm 13 --oem 3 -c tessedit_char_whitelist=0123456789')
                    
                    print(f'Pitcher: {pitcher} Pitch Count: {count}')
                    
                    cv2.imshow('Pitcher',cropFramePitcher)
                    cv2.imshow('Count',cropFrameCount)
                    
                    got+=1
                    t=cap.get(cv2.CAP_PROP_POS_MSEC)
                    t=int(np.floor(t/1000))
                    lst.append([got,pitcher,count,t])

                # pitching continues past recorded moment
                elif consecutive > 7:
                    cv2.rectangle(frame,top_left,(top_left_p[0],bottom_right_p[1]),(255,80,80), 1)
                    cv2.rectangle(frame,(bottom_right_p[0],top_left[1]),(bottom_right[0],bottom_right_p[1]),(80,80,255), 2)
                    cv2.rectangle(frame,top_left_p, bottom_right_p,(0,255,0), 1)
                    cv2.putText(frame, "pitch recorded", (1014,478), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,255,0),2)
                    
            # p is not being detected in correct location
            else:
                consecutive=0
        
        #infographic is not being detected
        else:
            cv2.rectangle(frame,(1014,498),(1014+width,498+height),(0,0,255), 3)
            cv2.putText(frame, "Not detected", (1014,478), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,0,255),2)
            count=0
            
            
         #Display the resulting frame
        cv2.imshow('frame', frame)
        
        # Exit if ESC pressed
        k = cv2.waitKey(1) & 0xff
        if k == 27 : 
            break
    else:
        break
        
    n= n+skip
    
# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()

df=pd.DataFrame(lst,columns =['Pitch number','Pitcher','Pitch Count','time'])
df['timestamp']=pd.to_timedelta(df['time'],unit='s')
df['begin']=df['timestamp']-datetime.timedelta(0,10)
df['end']=df['timestamp']+datetime.timedelta(0,1)
df
```

After working around, I actually went back to the previous version that detects the infographic and watches the pitch count p. Reason being that it quickly allowed me to set the vertecies of the next two rectangles I wanted to place. 

The Pitcher rectangle starts at the same point the infographic top left point does and ends at the same x position as the p rectangle top left, and the the y position of the p rectangle bottom right.

Similarly, the Count rectangle starts in the x position of the p rectangle bottom right, and the y position of the p rectangle top left and ends at the x position of the infographic bottom right and the y position of the p rectangle bottom right.

with these points I then crop out the regions of the frame and utilize tesseract to convert the image of text to a text string and store the info in a pandas dataframe.

I ran this program and observed the results. The program did fairly well at collecting the two sets of image text, but there were some errors. I attempt to correct this by using the tesseract config option, assigning the text to be register as single lines of characters, and giving each image relavent lists of characters to allow or not. For the Pitcher list, I was getting results that included special characters for instance 'Keuche{' by giving this tesseract a blacklist that excluded all special and numeric characters I hope to correct this. For the count image, they should only be numerical, so this config call had a whitelist of only numeric characters.

The program also over collected certain pitches. If something obscurred the p, that caused it to reset the continuous variable, it would reregister the same pitch count and make another, duplicate entry into the dataframe. The effect that causes the vast majority of entries is the p being removed and the mph being reported. This occurs promptly after the pitch. If a duplicate entry occurs with the same pitch count due to obstruction, it would be logged after the first true positive entry. So, by taking note of all duplicate entries and only keeping the first for each pitch count I should log only the true positives. This same logic could be used to relax the constraint of the p continuous variable as well, since multiple entries would be removed anyway after the true positve. This would possibly capture the sneaky first and last pitches of the inning described below.
