```python
import numpy as np
import pandas as pd

import vlc, pafy, time
import tkinter as tk
import tkinter.ttk as ttk
from PIL import Image, ImageTk,ImageOps
pd.options.mode.chained_assignment = None  # default='warn'

n=6962

path='C://Users/16142/Desktop/Astro/2018 season/'
df=pd.read_csv(f'{path}2018 season pitch log.csv')
df['Date'] = pd.to_datetime(df['Date'])

url = df['game_url'][n-1]
t=df['start'][n-1]
game=f"{str(df['Game'][n]).zfill(3)} {df['Date'][n].date()} {df['Opponent'][n]}"
spectro=f"{df['game_pitch'][n-1]} {df['Opponent'][n-1]} {df['pitcher'][n-1]} pitch #{df['pitch_count'][n-1]} - {df['pitch_type'][n-1]}.tiff"

class Screen(tk.Frame):

    '''
    Screen widget: Embedded video player from local or youtube
    '''

    def __init__(self, parent, *args, **kwargs):
        tk.Frame.__init__(self, parent, bg='black')
        self.parent = parent
        self.settings={'width':1020,'height':500}
        # Creating VLC player
        self.instance = vlc.Instance("prefer-insecure")
        self.player = self.instance.media_player_new()

    def GetHandle(self):
        # Getting frame ID
        return self.winfo_id()

    def Play(self, _source,t):
        # Function to start player from given source
        Media = self.instance.media_new(_source)
        Media.get_mrl()
        
        self.player.set_media(Media)
        self.player.set_hwnd(self.winfo_id())
        
        self.player.play()
        self.player.set_time(t*1000+3000)
        
        info.delete(1.0,tk.END)
        info.insert(tk.END,f'{df.iloc[n-1]}')
        
        btn_play.config(bg='green')
        #btn_play.after(17000,lambda: btn_play.config(bg='white'))
        #btn_play.after(17000,lambda: player.Pause())
        
    def Pause(self):
        self.player.pause()
        
    def Stop(self):
        self.player.stop()
        
    def Quit(self):
        self.player.stop()
        root.destroy()
        
def prev(df):
    global n
    global t
    n-=1
    t=df['start'][n-1]

    game=f"{str(df['Game'][n]).zfill(3)} {df['Date'][n].date()} {df['Opponent'][n]}"
    spectro=f"{df['game_pitch'][n-1]} {df['Opponent'][n-1]} {df['pitcher'][n-1]} pitch #{df['pitch_count'][n-1]} - {df['pitch_type'][n-1]}.tiff"

    image=Image.open(f'{path}spectrograms/{game}/{spectro}')
    image = ImageOps.fit(image,(1020,500))

    spectrogram.image = ImageTk.PhotoImage(image)
    spectrogram.config(image=spectrogram.image)
    
    info.delete(1.0,tk.END)
    info.insert(tk.END,f'{df.iloc[n-1]}')

    #player.Play(best.url,t)


def _next(df):
    global n
    global t

    n+=1
    url = df['game_url'][n-1]
    t=df['start'][n-1]

    game=f"{str(df['Game'][n]).zfill(3)} {df['Date'][n].date()} {df['Opponent'][n]}"
    spectro=f"{df['game_pitch'][n-1]} {df['Opponent'][n-1]} {df['pitcher'][n-1]} pitch #{df['pitch_count'][n-1]} - {df['pitch_type'][n-1]}.tiff"

    #video = pafy.new(url)
    #best = video.getbest()

    image=Image.open(f'{path}spectrograms/{game}/{spectro}')
    image = ImageOps.fit(image,(1020,500))

    spectrogram.image = ImageTk.PhotoImage(image)
    spectrogram.config(image=spectrogram.image)
    
    info.delete(1.0,tk.END)
    info.insert(tk.END,f'{df.iloc[n-1]}')

    #player.Play(best.url,t)

def redo(df):
    global url
    global best
    url = df['game_url'][n-1]
    video = pafy.new(url)
    best = video.getbest()
    
def no(df):

    df['whistle'][n-1]=0
    df['type'][n-1]='no'
    df.to_csv(f'{path}2018 season pitch log.csv',index=False)
    df=pd.read_csv(f'{path}2018 season pitch log.csv')
    df['Date'] = pd.to_datetime(df['Date'])
    info.delete(1.0,tk.END)
    info.insert(tk.END,f'{df.iloc[n-1]}')

def pwip(df):

    df['whistle'][n-1]=1
    df['type'][n-1]='pwip'
    df.to_csv(f'{path}2018 season pitch log.csv',index=False)
    df=pd.read_csv(f'{path}2018 season pitch log.csv')
    df['Date'] = pd.to_datetime(df['Date'])
    info.delete(1.0,tk.END)
    info.insert(tk.END,f'{df.iloc[n-1]}')

def charge(df):

    df['whistle'][n-1]=1
    df['type'][n-1]='charge'
    df.to_csv(f'{path}2018 season pitch log.csv',index=False)
    df=pd.read_csv(f'{path}2018 season pitch log.csv')
    df['Date'] = pd.to_datetime(df['Date'])
    info.delete(1.0,tk.END)
    info.insert(tk.END,f'{df.iloc[n-1]}')

def note(df):
    note=notes.get()
    df['notes'][n-1]=note
    df.to_csv(f'{path}2018 season pitch log.csv',index=False)
    df=pd.read_csv(f'{path}2018 season pitch log.csv')
    df['Date'] = pd.to_datetime(df['Date'])
    info.delete(1.0,tk.END)
    info.insert(tk.END,f'{df.iloc[n-1]}')
    
def restamp(df):
    global t
    t=notes.get()
    df['start'][n-1]=int(t)
    df['adjusted'][n-1]=(int(t)+20)
    df.to_csv(f'{path}2018 season pitch log.csv',index=False)
    df=pd.read_csv(f'{path}2018 season pitch log.csv')
    df['Date'] = pd.to_datetime(df['Date'])
    
video = pafy.new(url)
best = video.getbest()

root = tk.Tk()
root.title('Whistle Widget')
root.rowconfigure(1, weight=1)
root.columnconfigure(2, weight=1)

image=Image.open(f'{path}spectrograms/{game}/{spectro}')
image = ImageOps.fit(image,(1020,500))
image = ImageTk.PhotoImage(image)

info = tk.Text(root)
photo = tk.Frame(root,relief=tk.RAISED,bd=2)
buttons =tk.Frame(root,relief=tk.RAISED,bd=2)
player = Screen(root)
notes=tk.Entry(root)

btn_play = tk.Button(buttons, text="Play", command= lambda: player.Play(best.url,t))
btn_play.grid(row=0, column=0, sticky="nsew")
btn_play.bind_all('<Home>',lambda event : player.Play(best.url,t))

btn_pause = tk.Button(buttons, text="Pause", command=player.Pause)
btn_pause.grid(row=1, column=0, sticky="nsew")

btn_stop = tk.Button(buttons, text="Stop", command=player.Stop)
btn_stop.grid(row=2, column=0, sticky="nsew")

btn_prev = tk.Button(buttons, text="Prev", command= lambda : prev(df))
btn_prev.grid(row=3, column=0, sticky="nsew")
btn_prev.bind_all('<Prior>',lambda event : prev(df))

btn_next = tk.Button(buttons, text="Next", command=lambda : _next(df))
btn_next.grid(row=4, column=0, sticky="nsew")
btn_next.bind_all('<Next>',lambda event : _next(df))

btn_no = tk.Button(buttons, text='No',command=lambda: no(df))
btn_no.grid(row=5, column=0, sticky="nsew")
btn_no.bind_all('<F1>',lambda event : no(df))

btn_pwip = tk.Button(buttons, text='Pwip',command=lambda: pwip(df))
btn_pwip.grid(row=6, column=0, sticky="nsew")
btn_pwip.bind_all('<F2>',lambda event : pwip(df))

btn_charge = tk.Button(buttons, text='Charge',command=lambda: charge(df))
btn_charge.grid(row=7, column=0, sticky="nsew")
btn_charge.bind_all('<F3>',lambda event : charge(df))

btn_note = tk.Button(buttons, text='Note',command=lambda: note(df))
btn_note.grid(row=8, column=0, sticky="nsew")
btn_note.bind_all('<Return>',lambda event : note(df))

btn_time=tk.Button(buttons, text='Time',command=lambda: restamp(df))
btn_time.grid(row=9, column=0, sticky='nsew')

btn_url = tk.Button(buttons, text='Redo',command=lambda: redo(df))
btn_url.grid(row=10, column=0, sticky='nsew')

btn_quit = tk.Button(buttons, text='Quit',command=player.Quit,background='red')
btn_quit.grid(row=11,column=0, sticky='nsew')

spectrogram=tk.Label(photo,image=image)
spectrogram.grid(row=0, column=0, sticky="nsew")

buttons.grid(row=0,column=0, sticky='ns')
player.grid(column=1,row=0, sticky='nsew')
photo.grid(row=0,column=2, sticky='nsew')
info.grid(row=1,column=1, sticky='nsew',ipadx=10)
notes.grid(row=1,column=2,sticky='ew')

root.mainloop()
```
