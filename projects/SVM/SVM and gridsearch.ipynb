```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
%matplotlib inline

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
```

## Breast Cancer

```python
cancer =load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
cancer['target']
```

```python
X=df_feat
y=cancer['target']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid=GridSearchCV(SVC(),param_grid,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred=grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_pred))
print('\n')
print(classification_report(y_test,grid_pred))
```

## Pulsars

```python
df=pd.read_csv('pulsar_stars.csv')
```

```python
df.head()
```

```python
#columns too long

df.columns = ['IP MEAN', 'IP SD','IP KURTOSIS','IP SKEWNESS','DM-SNR MEAN','DM-SNR SD',
              'DM-SNR KURTOSIS', 'DM-SNR SKEWNESS','TARGET']
```

```python
# no missing values
df.isnull().sum()
```

```python
sns.boxplot(data=df,x='IP MEAN')
```

```python
sns.boxplot(data=df,x='IP SD')
```

```python
#many outliers, deal with them using SVMs
```

```python
#beautiful data with some pairs that have great correlation
sns.pairplot(data=df)
```

```python
X=df.drop(['TARGET'],axis=1)
y=df['TARGET']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
cols = X_train.columns
```

```python
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train,columns=[cols])
X_test=pd.DataFrame(X_test,columns=[cols])
```

```python
model=SVC()

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
param_grid=[{'C':[1,10,100,1000],'kernel':['linear']},
            {'C':[1,10,100,1000],'kernel':['rbf'], 'gamma':[0.001,0.01,0.1,0.5,1]},
            {'C':[1,10,100,1000],'kernel':['poly'],'degree':[2,3,4],'gamma':[0.001,0.01,0.1,]}]

grid=GridSearchCV(SVC(),param_grid,verbose=3)
grid.fit(X_train,y_train)
grid_pred=grid.predict(X_test)
```

```python
grid.best_params_
```

```python
print(accuracy_score(y_test,grid_pred))
print(confusion_matrix(y_test,grid_pred))
print(classification_report(y_test,grid_pred))
```
