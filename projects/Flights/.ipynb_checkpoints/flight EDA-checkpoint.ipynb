# Compare Airlines

- create 2 dataframe groupbys, 1 for the sum and one for the mean

```python
airline=flights.groupby('AIRLINE').sum()

airline=airline.reset_index()
airline.rename(columns={'AIRLINE':'IATA_CODE'},inplace=True)

airline_sum=pd.merge(airlines,airline, on='IATA_CODE')
del airline_sum['DAY_OF_WEEK']
airline_sum
```

```python
total=flights.groupby('AIRLINE').count()['ORIGIN_AIRPORT']
total=total.reset_index()

total.rename(columns={'AIRLINE':'IATA_CODE','ORIGIN_AIRPORT':'TOTAL_FLIGHTS'},inplace=True)
```

```python
airline_sum2=pd.merge(airline_sum,total,on='IATA_CODE')

airline_sum2.sort_values(by=['ARRIVAL_DELAY'],ascending=False,inplace=True)

airline_sum2.to_csv('Data/Airline_sums.csv',index=False)
```

```python
airlinem=flights.groupby('AIRLINE').mean()

airlinem=airlinem.reset_index()
airlinem.rename(columns={'AIRLINE':'IATA_CODE'},inplace=True)

airline_mean=pd.merge(airlines,airlinem, on='IATA_CODE')

airline_mean=airline_mean[['IATA_CODE']+['AIRLINE']+['DEPARTURE_DELAY']+['ARRIVAL_DELAY']]

airline_mean.to_csv('Data/Airline_means.csv',index=False)
```

```python
#set font for figure
font = {'weight' : 'normal',
        'size'   : 18}

matplotlib.rc('font', **font)
#arrange x axis ticks and labels of airlines, set width of bar graph bars
x=np.arange(len(airline_mean['IATA_CODE']))
width = 0.35

#set up figure, use width setting and x range to orient
fig, ax = plt.subplots(figsize=(15,10))
rects1= ax.bar(x-width/2,airline_mean['DEPARTURE_DELAY'],width,label='Departures')
rects2=ax.bar(x+width/2,airline_mean['ARRIVAL_DELAY'],width,label='Arrivals')

#labeling of axis and plot, call legend and grid lines
ax.set_xlabel('Airline',fontsize=20)
ax.set_ylabel('Average Delay in minutes',fontsize=20)
ax.set_title('Average Arrival and Departure Delay by Airline',fontsize=22)
ax.set_xticks(x)
ax.set_xticklabels(airline_mean['IATA_CODE'])
ax.legend(loc='upper right', facecolor='white',framealpha=1)
ax.xaxis.grid()
ax.yaxis.grid()
ax.set_axisbelow(True)


fig.tight_layout()
fig.show()
fig.savefig('EDA figures/avearge delays by airline.tif')
```

one take away from this first graph is that it appears that there is some delay baked into the arrival time. planes on average take off at a certain delay but arrive with less, excluding Hawaii Airlines... Be interesting if I can tease out that delay across all airlines. 

Pie chart that shows percent of total flights for each airline

```python
# Total flights 5819079
pie=airline_sum2.sort_values(by='TOTAL_FLIGHTS',ascending=False)

colors=['xkcd:azure','xkcd:red','xkcd:light cyan','xkcd:neon green','xkcd:rose','xkcd:violet','xkcd:chartreuse',
       'xkcd:pale purple','xkcd:coral','xkcd:pale orange','xkcd:bright pink','xkcd:spring green','xkcd:pale pink',
       'xkcd:rose']

explode=[0.4 if pie['TOTAL_FLIGHTS'].values[n]/5819079 < .04 else 0.0 for n in range(len(pie.values))]

fig, ax = plt.subplots(figsize=(14,10))

patches, texts, autotexts =ax.pie(pie['TOTAL_FLIGHTS'], explode=explode, labels=pie['IATA_CODE'],
                            wedgeprops={'linewidth': 1.0, 'edgecolor':'black'},colors=colors, autopct='%1.1f%%')


fp1=FontProperties(weight='normal',size=18)
fp2=FontProperties(weight='bold',size=15)
for n in range(len(pie.values)):
    texts[n].set_fontproperties(fp1)
    autotexts[n].set_fontproperties(fp2)
    
ax.axis('equal')
ax.set_title('Total Flight percentage by Airline',fontsize=22,loc='right')
leg=ax.legend(pie['AIRLINE'],fontsize=15,bbox_to_anchor=(1,0.8))
plt.tight_layout()
plt.show()
fig.savefig('EDA figures/Figure 2 - Total Flight percentage by Airline.tif',bbox_extra_artists=(leg,), bbox_inches='tight')
```

```python
MQ_DATE=MQ.groupby('FULLDATE').sum()
MQ_COUNT=MQ.groupby('FULLDATE').count()['AIRLINE'].reset_index()
MQ_DATE=pd.merge(MQ_DATE,MQ_COUNT, on='FULLDATE')
MQ_DATE.rename(columns={'AIRLINE':'DAY_TOTAL'},inplace=True)

WN_DATE=WN.groupby('FULLDATE').sum()
WN_COUNT=WN.groupby('FULLDATE').count()['AIRLINE'].reset_index()
WN_DATE=pd.merge(WN_DATE,WN_COUNT, on='FULLDATE')
WN_DATE.rename(columns={'AIRLINE':'DAY_TOTAL'},inplace=True)

WN_DATE.index=WN_DATE['FULLDATE']
WN_DATES=WN_DATE.resample('M').sum()

MQ_DATE.index=MQ_DATE['FULLDATE']
MQ_DATES=MQ_DATE.resample('M').sum()
```

```python
airp2=flights.groupby(['AIRLINE','ORIGIN_AIRPORT']).count()
airp4=flights.groupby(['AIRLINE','ORIGIN_AIRPORT']).sum()

airp3=airp2['DESTINATION_AIRPORT']
airp3=pd.DataFrame(data=airp3)

airp3.rename(columns={'DESTINATION_AIRPORT':'AIRPORT_COUNT'},inplace=True)

airp4.reset_index(inplace=True)

airp3.reset_index(inplace=True)

airp5=pd.merge(airp4,airp3,on=['AIRLINE', 'ORIGIN_AIRPORT'])

airports2=airports.rename(columns={'IATA_CODE':'ORIGIN_AIRPORT'})

airline_portusage=pd.merge(airp5,airports2, on='ORIGIN_AIRPORT')
airline_portusage.sort_values(['AIRLINE','ORIGIN_AIRPORT'],inplace=True,ascending=True)

airline_portusage = airline_portusage[['AIRLINE',
                                         'ORIGIN_AIRPORT',
                                         'AIRPORT',
                                         'CITY',
                                         'STATE',
                                         'COUNTRY',
                                         'LATITUDE',
                                         'LONGITUDE',
                                         'DEPARTURE_DELAY',
                                         'ARRIVAL_DELAY',
                                         'DIVERTED',
                                         'CANCELLED',
                                         'ARRIVED_LATE',
                                         'DEPARTED_LATE',
                                         'AIRPORT_COUNT'
                                         ]]

airline_portusage.to_csv('Data/Airline_portusage.csv',index=True)
airline_portusage.shape
```

I now have a by airline by airport breakdown of information. I want to be able to see usage graphically, especially for MQ, use plotly.

```python
def Airport_Map(CODE):
    air=airline_portusage[airline_portusage['AIRLINE'].values==CODE]
    fig.data=[]

    size=list(air['AIRPORT_COUNT'])
    air_text=air[['AIRPORT']+['AIRPORT_COUNT']]
    fig.add_trace(go.Scattergeo(locationmode='USA-states', lon=air['LONGITUDE'], lat=air['LATITUDE'],text=air_text,
                                    marker = dict(size=air['AIRPORT_COUNT'], opacity=1,
                                                  color='red', sizemode='area',sizemin=3,sizeref=2.*max(size)/(40.**2))))

    fig.update_layout(title_text = f'2015 {CODE} Airline airport usage population', showlegend=False,
                      geo=dict(scope='usa',landcolor='rgb(217,217,217)'))
    fig.write_image(f'EDA figures/figure 7 - {CODE} geoscatter.jpeg')
    return fig.show()
```
