```python
def time_standard(str1):
    dt_time=0
    if str1=='nan':
        dt_time=None
    elif len(str1)==3:
        str1='00-0'+str1[0]
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif len(str1)==4:
        str1='00-'+str1[0:2]
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif len(str1)==5:
        str1='0'+str1[0]+'-'+str1[1:3]
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif str1=='2400.0':
        str1='00-00'
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif len(str1)==6:
        str1=str1[0:2]+'-'+str1[2:4]
        dt_time=datetime.strptime(str1, '%H-%M').time()
        
    return dt_time

def sched_standard(str1):
    dt_time=0
    if str1=='nan':
        dt_time=None
    elif len(str1)==1:
        str1='00-0'+str1[0]
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif len(str1)==2:
        str1='00-'+str1[0:2]
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif len(str1)==3:
        str1='0'+str1[0]+'-'+str1[1:3]
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif str1=='2400':
        str1='00-00'
        dt_time=datetime.strptime(str1, '%H-%M').time()
    elif len(str1)==4:
        str1=str1[0:2]+'-'+str1[2:4]
        dt_time=datetime.strptime(str1, '%H-%M').time()
        
    return dt_time
```

We can do this for each state. find total airports, international airports, and then find the 'density' of international airports and add it to the dataframe as a new column. We have a list of total airports by state already. Created a dataframe of all the international airports. count shows that 10 states and territories don't have international airports. which ones? Will find out when their international density is 0.

```python
int_air=airports[airports['AIRPORT'].str.contains('International')]['STATE'].value_counts()
```

```python
d={}
STATE_TOT=states.reset_index()
STATE_INT=int_air.reset_index()
for index,total2 in STATE_INT.iterrows():
    for index,total1 in STATE_TOT.iterrows():
        if total1[0]==total2[0]:
            density=total2[1]/total1[1]
            d.update({total1[0]:density})

df=pd.Series(d, name='DENSITY')
df.index.name = 'index'
DENSITY=df.reset_index()
DEN=pd.merge(STATE_TOT,STATE_INT,on='index',how='outer')
DENSITY=pd.merge(DEN,DENSITY,on='index',how='outer')
DENSITY.fillna(0)
DENSITY.rename(columns={'index':'STATE','STATE_x':'TOTAL','STATE_y':'INTERNATIONAL','DENSITY':'DENSITY'},inplace=True)

DENSITY.sort_values(['DENSITY'],ascending=False,inplace=True)
```

This is good but not quite good enough for me. Small states and territories are getting weighted to heavily. We are going to add one more column, that weights the density by the total number airports.

```python
#airports.count()
#322

DENSITY['TOT_DEN']=DENSITY['TOTAL'].apply(lambda x: x/322)
DENSITY['WEIGHTED_DENSITY']=DENSITY.apply(lambda x: x['DENSITY']*x['TOT_DEN']*100,axis=1)
DENSITY.sort_values(['WEIGHTED_DENSITY'],ascending=False,inplace=True)
DENSITY.fillna(0)
```

```python
sample=flights[0::1000]
```

```python
# initiate empty lists to be converted to additional columns
n=0
DEPART=[]
ARRIVE=[]
SCHED_DEPART=[]
SCHED_ARRIVE=[]
DELAY_DEPART=[]
ARRLATE=[]
DEPLATE=[]
FULLDATE=[]

#interate through each row and do the operations I want
for index,row in sample.iterrows():
    
    #create a string and change it to a datetime for the fulldate
    fulldate=f'{row[1]}-{row[2]}-15'
    dt_date=datetime.strptime(fulldate, '%m-%d-%y').date()
    FULLDATE.append(dt_date)
    
    #pull the subpar data from the depart and arrival columns
    #which used military time in float format and convert them to date time
    SCHED_DEPART.append(sched_standard(f'{row[9]}'))
    DEPART.append(time_standard(f'{row[10]}'))
    SCHED_ARRIVE.append(sched_standard(f'{row[20]}'))
    ARRIVE.append(time_standard(f'{row[21]}'))
    
    #Build categorical columns that report lateness of arrival or departure for t>= 20 minutes
    if row[11]>= 20:
        DEPLATE.append(int(1))
    else:
        DEPLATE.append(int(0))
        
    if row[22]>=20:
        ARRLATE.append(int(1))
    else:
        ARRLATE.append(int(0))
    
    #give me a benchmark on how far along the processing is
    if n%10000==0:
        print('at entry',n)
    if n%100000==0:
        clear_output()
    n+=1

#put all the new columns onto the dataframe
sample['FULLDATE']=FULLDATE
sample['SCHED_DEPART']=SCHED_DEPART
sample['DEPART']=DEPART
sample['SCHED_ARRIVE']=SCHED_ARRIVE
sample['ARRIVE']=ARRIVE
sample['ARRIVED_LATE']=ARRLATE
sample['DEPARTED_LATE']=DEPLATE

#reorganize the columns to my prefered order, and drop the columns we won't be using
sample2=sample[['AIRLINE']+['ORIGIN_AIRPORT']+['DESTINATION_AIRPORT']+['TAIL_NUMBER']+['FULLDATE']+['DAY_OF_WEEK']
           +['SCHED_DEPART']+['DEPART']+['DEPARTURE_DELAY']+['SCHED_ARRIVE']+['ARRIVE']+['ARRIVAL_DELAY']
           +[c for c in sample if c not in ['YEAR','MONTH','DAY','FULLDATE','SCHED_DEPART','DEPART','DEPARTURE_DELAY',
           'ARRIVAL_DELAY','SCHED_ARRIVE','DAY_OF_WEEK','SCHEDULED_ARRIVAL','ARRIVAL_TIME','SCHEDULED_DEPARTURE',
           'DEPARTURE_TIME','ARRIVE','TAXI_OUT','WHEELS_OFF', 'SCHEDULED_TIME', 'ELAPSED_TIME', 'AIR_TIME', 'DISTANCE',
           'WHEELS_ON', 'TAXI_IN','CANCELLATION_REASON', 'AIR_SYSTEM_DELAY','SECURITY_DELAY',
           'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY','FLIGHT_NUMBER','WEATHER_DELAY','AIRLINE', 'TAIL_NUMBER',
           'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT']]]
```

The result of this routine is a nice clean dataset with some known issues. The columns are reorganized, the time columns are standardized, and a couple extra categorical columns are added. The columns that we won't be interested in are dropped and we keep in mind the issue with the October airport codes.
