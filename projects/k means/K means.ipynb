```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)

datatest = make_blobs(n_samples=200,n_features=3,centers=5,cluster_std=1.2,random_state=101)
```

```python
type(data[0])
```

```python
data[0]
```

```python
plt.scatter(datatest[0][:,0],datatest[0][:,1],c=datatest[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans= KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
data[1]
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
kmeans2= KMeans(n_clusters=2)
```

```python
kmeans2.fit(data[0])
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans2.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
