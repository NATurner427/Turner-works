```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation
import matplotlib as mpl

from scipy.spatial.distance import pdist, squareform
import scipy.integrate as integrate
```

```python
class Ballistic:
    
    def __init__(self,initial,bounds=[-3,3,-2,2],size=3,mass=1,fps=24,N=2):
        
        self.initial=np.asarray(initial,dtype=float)
        self.mass=mass*np.ones(self.initial.shape[0])
        self.size=size
        self.state=self.initial.copy()
        self.time=0
        self.bounds=bounds
    
    def delta(self,dt):
        
        #keep track of time
        self.time +=dt
        
        # add the current position of balls and the velocity of those balls multiplied by delta time to get new position
        self.state[:, :2] = self.state[:, :2] + self.state[:, 2:]*dt
        
        # find new distance between each ball, and every other ball
        D=squareform(pdist(self.state[:,:2]))
        
        # check to see if any collisions occured
        ind1,ind2 = np.where(D< .8*self.size)
        hit=(ind1<ind2)
        ind1=ind1[hit]
        ind2=ind2[hit]
        
        #calculate new velocities of collisions. 
        #If a collision occured pull masses and do physics if not, whole loop is skipped.
        #picks out the pairs of collisions that occurered
        for i1,i2 in zip(ind1,ind2):
            m_1=self.mass[i1]
            m_2=self.mass[i2]
            
            #grab position of colliding balls
            r_1=self.state[i1, :2]
            r_2=self.state[i2, :2]
            
            #grab velocity of colliding balls
            v_1=self.state[i1, 2:]
            v_2=self.state[i2, 2:]
            
            # relative position and velocity
            r_rel= r_2 - r_1
            v_rel= v_2 - v_1
            
            # momentum vector of the center of mass of the two body collision
            p_cm = (m_1*v_1 + m_2*v_2)/(m_1+m_2)
            
            # velocity of ball that has collided is reflected across the position vector
            rr_rel=np.dot(r_rel,r_rel)
            vr_rel=np.dot(v_rel,r_rel)
            v_rel=2*r_rel* vr_rel/rr_rel -v_rel
            
            #set new velocities with some energy loss
            self.state[i1,2:]=(p_cm +v_rel *(m_2/(m_1+m_2)))*.9
            self.state[i2,2:]=(p_cm -v_rel *(m_1/(m_2+m_1)))*.9
            
        #boundary condition: reflection
        # if any x positions are approaching left of box
        crossed_x1 = self.state[:,0] < self.bounds[0] +self.size
        
        #if any x positions are approaching right of box
        crossed_x2 = self.state[:,0] > self.bounds[1] -self.size
        
        #if any y positions are approaching bottom of the box
        crossed_y1 = self.state[:,1] < self.bounds[2] +self.size
        
        #if any y positions are approaching top of box
        crossed_y2 = self.state[:,1] > self.bounds[3] -self.size
        
        # pull the ones we found and set there positions to be within box
        self.state[crossed_x1, 0] = self.bounds[0] + self.size
        self.state[crossed_x2, 0] = self.bounds[1] - self.size
        self.state[crossed_y1, 1] = self.bounds[2] + self.size
        self.state[crossed_y2, 1] = self.bounds[3] - self.size
        
        # if ball hit side in the x direction, reflect its velocity with some energy loss
        self.state[crossed_x1 | crossed_x2, 2] *= -.9
        
        #if ball hit side in the y direction, reflect its velocity with some energy loss
        self.state[crossed_y1 | crossed_y2, 3] *= -.9
            
```

## set up an array for my ballistics

use numpy array that has N rows and 4 columns. the first column will x position, 2nd column y position, 3rd column x velocity and 4th column is y velocity.

```python
#frames per second
fps=24

#seconds*frames per second to get total frames
frames= 20*fps

#number of ballistics
N=50
size=5

# random starting position. np.random makes numbers between 0 and 1. so subtract .5 and multiply  by 2 to get 1 to -1
initial=(np.random.random((N, 4)) - 0.5)*2

#limits of box
bx=3
by=2
bounds=[-bx,bx,-by,by]

# scale up randomized position values by dimensions of box I want to bounce around in. select first column and scale up
#so that x value (-.5 to .5) will fit box.
initial[:,0] *= bx

#scale up y positions in 2nd column same way 
initial[:,1]*=by

#intialize class
ballistic=Ballistic(initial,size=size/100,fps=fps,N=N,mass=10)
dt=1./fps

#set up figure
fig = plt.figure()
ax=fig.add_subplot(xlim=(-3.2,3.2), ylim=(-2.4,2.4))
plt.title(f'{N} ball animation test')
ax.set_axis_off()

# positions
point= ax.scatter(initial[:,0],initial[:, 1],s=size*3)

# box is the box edge
box = plt.Rectangle(bounds[::2],
                     bounds[1] - bounds[0],
                     bounds[3] - bounds[2],
                     ec='none', lw=2, fc='none')
ax.add_patch(box)

# initial positions of balls and box
def init():
    point.set_offsets(initial[:,:2])
    box.set_edgecolor('none')
    return point, box

# step of calculation and animation
def animate(i):
    ballistic.delta(dt)
    box.set_edgecolor('k')
    point.set_offsets(ballistic.state[:, :2])
    
    
    return point, box

ani = animation.FuncAnimation(fig,animate, frames=frames, interval=10,blit=True)

ani.save(f'ballistics/{N} ball test.mp4', fps=fps, extra_args=['-vcodec', 'libx264'])
plt.show()
```
