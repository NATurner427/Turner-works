### simple exponential growth equation

$$F(t)=P_o*e^{ \displaystyle (r*t)}$$

For a given initial population, $P_o$ and rate, $r$, what is the population at some time=$t$ later?

This kind of equation was apply in your highschool or college math courses for petri dish cultures, intreset rates and population growths.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.font_manager import FontProperties
import math
%matplotlib inline

# make font legible size
font={'size':18}

matplotlib.rc('font',**font)
```

```python
#create a t variable for our order pairs
t=np.linspace(0,100,101)

# chose a rate and initial population
r=.06
P_o=1

#the math
Ft=P_o*np.exp(r*t)

# plot and make good looking
fig, ax = plt.subplots(figsize=(10,8))
line, = ax.plot(t,Ft,'ro')
ax.set_title('Exponential Growth')
ax.set_ylabel('Population')
ax.set_xlabel('time')
ax.set_xlim(0,101)
ax.set_ylim(0,410)
ax.xaxis.grid()
ax.yaxis.grid()
line.set_label('population')
ax.legend()

plt.tight_layout()
np.floor(Ft[-1])
```

 starting with a population of 1 with a rate of 6% or .06 after 100 'times' (minutes, hours, days whatever is relevant to the rate) later you have a population of 403.

how much population is added everytime?

```python
#find the difference between sequential days
F_del=np.diff(Ft)

#one less time because we are doing time 2-time 1, time 3 - time 2... get to time 100 - time 99. 
#no final data point for time 101 - time 100, because we didn't have a time 101.
t=np.linspace(0,99,100)

fig, ax = plt.subplots(figsize=(10,8))
line, = ax.plot(t,F_del,'bo')
ax.set_title('Exponential Growth Difference')
ax.set_ylabel('Population increase')
ax.set_xlabel('time')
ax.set_xlim(0,100)
ax.set_ylim(0,40)
ax.xaxis.grid()
ax.yaxis.grid()
line.set_label('population difference')
ax.legend()

plt.tight_layout()

print('number of new poplulation, rounded down, by unit time')
print(np.floor(F_del))
```

by the last few times, 20+ new population is being added each time.

### simple exponential decay equation

$$F(t)=P_o*e^{ \displaystyle (-r*t)}$$

For a given initial population, $P_o$ and rate, $r$, what is the population at some time=$t$ later?

the difference is now the rate is negative, so the population will be diminishing as time goes on. If we make the initial population equal to the final population in the exponential growth case we will 'flip' the figure. Given the same initial conditions as the exponential growth figures above the result will be them 'flipping'.

```python
#create a t variable for our order pairs
t=np.linspace(0,100,101)

# chose a rate and initial population
r=.06
P_o=403

#the math
Ft=P_o*np.exp((-r)*t)

# plot and make good looking
fig, ax = plt.subplots(figsize=(10,8))
line, = ax.plot(t,Ft,'ro')
ax.set_title('Exponential Decay')
ax.set_ylabel('Population')
ax.set_xlabel('time')
ax.set_xlim(0,101)
ax.set_ylim(0,410)
ax.xaxis.grid()
ax.yaxis.grid()
line.set_label('population')
ax.legend()

plt.tight_layout()
np.floor(Ft[-1])
```

```python
#find the difference between sequential days
F_del=np.diff(Ft)

#one less time because we are doing time 2-time 1, time 3 - time 2... get to time 100 - time 99. 
#no final data point for time 101 - time 100, because we didn't have a time 101.
t=np.linspace(0,99,100)

fig, ax = plt.subplots(figsize=(10,8))
line, = ax.plot(t,np.abs(F_del),'bo')
ax.set_title('Exponential decay Difference')
ax.set_ylabel('Population decrease')
ax.set_xlabel('time')
ax.set_xlim(0,100)
ax.set_ylim(0,40)
ax.xaxis.grid()
ax.yaxis.grid()
line.set_label('absolute population difference')
ax.legend()

plt.tight_layout()

print('number of new poplulation, rounded down, by unit time')
print(np.floor(np.abs(F_del)))
```

### exponential infection

Some things can sustain exponential growth indefinitely, interest will always accrue. But when considered in more detail others can not sustain such growth forever. Human population growth can be modeled as an exponential growth, but at some point the things that facilite that exponential growth will run out. No more resources like food or habitable land. So realalistically there need to be limits on these growths. What those limits are and how they are described mathematically will be different for each case. 

Lets start to focus in on infection to consider limits.

In this case we have an initial population of whatever the human population is being considered. This calls for a chosen simplification. In reality, the population and therefore the invectors and invectees is fluid. Travelers and vistors are a common thing, if they weren't, the infection we are interested in would never have left its initial outbreak location. So we make the limitation that no one is coming or going in the population, at least for now. so if we only have so many humans in consideration, how will the number of infected change over time?

lets say our inital population is 999 identical people. then an infected 999th person arrives. On the first day, he infects 2 people. Now there are 3 invected people. On day 2, each of those 3 people infect 2 people, thats 6 newly infected people plus the 3 that did the invecting for a total of 9. Going this way, how soon before all 1000 people will be infected? We have initial unaffected $P_o = 1000$. but now that population is not the thing growing, whats growing is the invectors, call that x. on day $t=0,1,2,...$  $x(t)=1,3,9...$ and therefore the remaining uninvected population is $F(t)=999,997,991...$ 

We can start to write the equation as  $$F(t)=P_o - x(t)$$

But now the job is describing how x depends on t. For this case, I can see the pattern. $x(t)=3^{\displaystyle t}$ gives us the correct set of values.  

and so $$F(t)=1000-3^{\displaystyle t}$$

```python
xr=8

t=np.arange(0,xr)

P_o=1000
x=3

Ft=P_o - x**t

# plot and make good looking
fig, ax = plt.subplots(figsize=(10,8))
line, = ax.plot(t,Ft,'g-')
ax.set_title('Uninfected Population over time')
ax.set_ylabel('Population')
ax.set_xlabel('time')
ax.set_xlim(0,xr)
ax.set_ylim(-5,1050)
ax.xaxis.grid()
ax.yaxis.grid()
line.set_label('population')
ax.legend()

plt.tight_layout()

print('it takes 7 days to reach a negative number')
print(Ft)
```

what happens when we change how many people get infected each day?

if the 1000th person infected 3 people, and they each infected 3 people what do we get? Now we have x(t)=1,4,16...

then the equation becomes $$F(t)=1000-4^{\displaystyle t}$$

```python
xr=8

t=np.arange(0,xr)

P_o=1000
x=4

Ft=P_o - x**t

# plot and make good looking
fig, ax = plt.subplots(figsize=(10,8))
line, = ax.plot(t,Ft,'g-')
ax.set_title('Uninfected Population over time')
ax.set_ylabel('Population')
ax.set_xlabel('time')
ax.set_xlim(0,xr)
ax.set_ylim(-5,1050)
ax.xaxis.grid()
ax.yaxis.grid()
line.set_label('uninfected population')
ax.legend()

plt.tight_layout()

print('it takes 5 days to reach a negative number')
print(Ft)
```

One thing we need to do is 'shut off' the infection. once 1000 people are infected, more people can't get infected. But my expression as it is written keeps plummeting to negative infinity. In the same vein, possibly nievly, there should be a certain spatial factor that needs to be considered. For example, look at the above examples day 4 population. there are 744 uninfected people, and therefore 256 infected people. If we picture them all in a big group, the 256 would be clustered together,zombie like and centered around the initial infector. Just with our imagination looking at this mental picture, it seems unreasonable that each of these 256 people could spatially 'reach' the remaining. There would be pushed out 'runt of the litter situations' when one infector can't get to anybody who isn't already infected. This would blunt the rate of infection. Even scaling up to an actual city of sick and healthy there would be those who would be easily accessed and those who weren't. There are door to door salesmen but there are also do not solicite hermits.

We need to adjust the x(t) function to account for this.

look back to the exponential decay figure, it went from an initial population to 0 if I want x(t) to start unaffected as 1 and then diminsh to 0 I just need to use a factor in the form of $e^{-rt}$.

when $t = 0, e^{-rt}=1$ and as $t \to \infty , e^{-rt} \to 0$

Then, how quickly the factor goes to 0 depends on the spreading rate r. if r is large the exponetial quickly goes to 0 and the infection dies out. If r is small, the exponential slowly goes to 0 and the infection spreads quickly. I want it so that a large r leads to a quick infection so I'm going to flip the r over into the denomiator.

$$e^{-t/r}$$

Going one step further, I think I want to use the fermi-dirac distribution. at t=negative infinity its 1, at t=0 its .5 and at t= positive infinity its 0.

By adjusting paramters I can change where the switch from 1 to 0 is and how strongly it switches.

to generalize this, we can rearrange and relabel my equation 

$$F(t)=P_o - x(t)$$

lets call $F(t)$, the number of uninfected, or *susceptible* people, $S(t)$, and $x(t)$, the number of *infected* people, $I(t)$, and $P_o$, the total *number* of people, $N$.

$$S(t)=N - I(t)$$

Now lets include a new variable. We have not yet infected $S(t)$, and $infected I(t)$ but theres also post infected *recovered* people $R(t)$. Then eventually all infected people (who didn't perish) would eventually become recovered people, and no susceptible people can become recovered people, they have to be infected people first.

So the equation becomes

$$S(t)=N - I(t)-R(t)$$

or, rearanged 

$$S(t)+I(t)+R(t)=N$$

and we have a flow relation

$$S(t) \to I(t) \to R(t)$$

We divide the whole equation  by N so we deal with fractions of the total population

$$S(t)/N+I(t)/N+R(t)/N=N/N \to$$

$$s(t)+i(t)+r(t)=1~~~(1)$$


### rate equations:

the change in the number of susceptible people $S(t)$ over time to infected people is a rate that can de written differentially as $\cfrac{dS}{dt}$ This rate describes how quickly $S(t)$ people become $I(t)$ people. it depends on how many contacts $S(t)$ people have had with $I(t)$ pepople, called $b$ and the fraction of the population that is still suspectible $s(t)$

$$\cfrac{dS}{dt}=-b~s(t)~I(t)$$

or dividing both sides by $N$

$$\cfrac{ds}{dt}=-b~s(t)~i(t)$$

Now the rate that infected people become recovered people can also be described differentially. 

$$\cfrac{dR}{dt}=k~I(t)$$

where k is the fraction of infected people that recover each day. again divided through by N.

$$\cfrac{dr}{dt}=k~i(t)$$

Now take the derivative with respect to time of equation $(1)$ and you get

$$\cfrac{d}{dt}(s(t)+i(t)+r(t))=\cfrac{d}{dt}1$$

$$\cfrac{ds}{dt}+\cfrac{di}{dt}+\cfrac{dr}{dt}=\cfrac{d}{dt}1$$

the derivative of the right side is 0 and we plug in the two rate equations to get

$$-b~s(t)~i(t)+\cfrac{di}{dt}+k~i(t)=0$$

reararanged to be 

$$\cfrac{di}{dt}=b~s(t)~i(t)-k~i(t)$$


then the SIR model is

$$\cfrac{ds}{dt}=-b~s(t)~i(t)$$

$$\cfrac{dr}{dt}=k~i(t)$$

$$\cfrac{di}{dt}=b~s(t)~i(t)-k~i(t)$$

where there are initial conditions $S(0)$, $I(0)$, $R(0)$ and $N$, contacts $b$ and recover rate $k$.

Lets do an example I found on the internet based on the hong long flu in New York City in 1960. There was no initial immunity to this so $N=S(0)=7.9x10e6$, $R(0)=0$ and assume only 10 initial infectors $I(0)=10$

then the fractions are

$$s(0)=1$$
$$r(0)=0$$
$$i(0)=1/7.9x10e6=1.27x10e-6$$

and guess at $b=.5$ and $k=.333$

### Euler's Method for Systems

Almost done talking. This is a way to use initial conditions to build our functions. 

we can piecemeal step through our 3 equations by taking the preceding value of each function and add to it the slope of the previous step multiplied with the step in dependent variable.

$$y_n=y_{n-1}+slope_{n-1}*delta~t$$

Or for our equations 

$$s_n=s_{n-1}+(-b~s_{n-1}~i_{n-1})dt$$

$$r_n=r_{n-1}+(k~i_{n-1})dt$$

$$i_n=i_{n-1}+(b~s_{n-1}~i_{n-1}-k~i_{n-1})dt$$

time to write a loop that steps through t to calculate these for the hong long flu.

```python
def PeakSick(l):
    max_val = max(l)
    max_idx = l.index(max_val)/100
    return print(f'peak sickness occured on day {np.floor(max_idx)} with {np.floor(max_val)} infected\n')


def Infection(N,S_o,I_o,R_o,days,b,k,loca,infec):
    s_o=S_o/N
    i_o=I_o/N
    r_o=R_o/N
    Ro=b/k
    check=s_o*Ro
    print(f'Basic Repoduction number is {Ro:.2f} and susceptible population fraction is {s_o:.2f}')
    if check >= 1:
        print(f'{check:.2f} is over 1: Endemic disease')
    elif check <1:
        print(f'{check:.2f} is under 1: relatively ok')
    # lists to be filled with s r and i functions
    s=[]
    r=[]
    i=[]

    #make time range in days with a set increment size
    day_o=0
    increment=days*100
    time=np.linspace(day_o,days,increment+1)
    #step size
    step=days/(increment)

    #loop
    while day_o < days:

        #get first values using initial conditions
        if day_o==0:
            s.append(s_o)
            r.append(r_o)
            i.append(i_o)

        else:
            s_n=(s[-1])-b*(s[-1])*(i[-1])*step
            r_n=(r[-1])+k*(i[-1])*step
            i_n=(i[-1])+(b*s[-1]*i[-1]-k*i[-1])*step
            s.append(s_n)
            r.append(r_n)
            i.append(i_n)

        day_o+=step
    #get values in terms of population
    S=[x*N for x in s]
    I=[x*N for x in i]
    R=[x*N for x in r]

    fig, ax = plt.subplots(figsize=(10,8))
    line, = ax.plot(time,S,'r-')
    line2,=ax.plot(time,I,'b-')
    line3,=ax.plot(time,R,'g-')
    ax.set_title(f'{loca} {infec} Infection over time')
    ax.set_ylabel('Population')
    ax.set_xlabel('Days')
    ax.set_xlim(0,days)
    #ax.set_ylim(-5,1050)
    ax.xaxis.grid()
    ax.yaxis.grid()
    line.set_label('Susceptible')
    line2.set_label('Infected')
    line3.set_label('Recovered')
    ax.legend()

    plt.tight_layout()
    
    PeakSick(I)
    return s,S,r,R,i,I

def Infection_w(N,S_o,I_o,R_o,days,b,k,loca,infec):
    s_o=S_o/N
    i_o=I_o/N
    r_o=R_o/N
    Ro=b/k
    check=s_o*Ro
    print(f'Basic Repoduction number is {Ro:.2f} and susceptible population fraction is {s_o:.2f}')
    if check >= 1:
        print(f'{check:.2f} is over 1: Endemic disease')
    elif check <1:
        print(f'{check:.2f} is under 1: relatively ok')
    # lists to be filled with s r and i functions
    s=[]
    r=[]
    i=[]

    #make time range in days with a set increment size
    day_o=0
    increment=days*100
    time=np.linspace(day_o,days,increment+1)
    #step size
    step=days/(increment)

    #loop
    while day_o < days:

        #get first values using initial conditions
        if day_o==0:
            s.append(s_o)
            r.append(r_o)
            i.append(i_o)

        else:
            s_n=(s[-1])-b*(s[-1])*(i[-1])*step
            r_n=(r[-1])+k*(i[-1])*step
            i_n=(i[-1])+(b*s[-1]*i[-1]-k*i[-1])*step
            s.append(s_n)
            r.append(r_n)
            i.append(i_n)

        day_o+=step
    #get values in terms of population
    S=[x*N for x in s]
    I=[x*N for x in i]
    R=[x*N for x in r]

    fig, ax = plt.subplots(figsize=(10,8))
    line, = ax.plot(time,s,'r-')
    line2,=ax.plot(time,i,'b-')
    line3,=ax.plot(time,r,'g-')
    ax.set_title(f'{loca} {infec} Infection over time')
    ax.set_ylabel('Population')
    ax.set_xlabel('Days')
    ax.set_xlim(0,days)
    #ax.set_ylim(-5,1050)
    ax.xaxis.grid()
    ax.yaxis.grid()
    line.set_label('Susceptible')
    line2.set_label('Infected')
    line3.set_label('Recovered')
    ax.legend()

    plt.tight_layout()
    
    PeakSick(I)
    return s,S,r,R,i,I
```

```python
#population
N_u=COMA.loc[COMA['DESIGNATION']=='U']['N'].sum()
N_s=COMA.loc[COMA['DESIGNATION']=='S']['N'].sum()
N_r=COMA.loc[COMA['DESIGNATION']=='R']['N'].sum()
N=N_u+N_s+N_r

#initial infected 
I_o=1
i_o=I_o/N

#initial suceptible
S_o=N-I_o

s_o=S_o/N


#initial recovered
R_o=0
r_o=R_o/N

#days out
days=250

#make time range in days with a set increment size
day_o=0
increment=days*24
time=np.linspace(day_o,days,increment+1)
#step size
step=days/(increment)

#contamination

b=.673

#recovery
k=.2693

#transit parameter
trav=0.0

#mitigate day in days
mitigate=0
off = 0
#mitigate effectiveness 1 off 0 ideal
m=1
b_m=b*m

Ro=b/k
Rom=b_m/k
check=s_o*Ro
checkm=s_o*Rom
print(f'Basic Repoduction number is {Ro:.2f} and susceptible population fraction is {s_o:.2f}')
if check >= 1:
    print(f'{check:.2f} is over 1: Endemic disease')
    if checkm >= 1:
        print(f'{checkm:.2f} is over 1 even after mitgation: Still endemic disease') 
        
    elif checkm < 1:
        print(f'{checkm:.2f} is now under 1 crisis averted') 
        
elif check <1:
    print(f'{check:.2f} is under 1: ok')
# lists to be filled with s r and i functions
s=[]
i=[]
r=[]

#loop
while day_o < days:

    #get first values using initial conditions
    if day_o==0:
        
        
        s.append(s_o)
      
        i.append(i_o)
      
        r.append(r_o)
        
    else:
        if day_o < mitigate or day_o > off:
            s_n=(s[-1])-b*(s[-1])*(i[-1])*step
            
            s.append(s_n)

            r_n=(r[-1])+k*(i[-1])*step
            
            r.append(r_n)

            i_n=(i[-1])+(b*s[-1]*i[-1]-k*i[-1])*step
            
            i.append(i_n)
        
        elif day_o >= mitigate and day_o <= off:
            
            s_n=(s[-1])-b_m*(s[-1])*(i[-1])*step
            
            s.append(s_n)

            r_n=(r[-1])+k*(i[-1])*step
            
            r.append(r_n)

            i_n=(i[-1])+(b_m*s[-1]*i[-1]-k*i[-1])*step
            
            i.append(i_n)


    day_o+=step
#get values in terms of population
S=[x*N for x in s]

I=[x*N for x in i]

R=[x*N for x in r]

fig, ax = plt.subplots(figsize=(12,10))
line_s, = ax.plot(time,s,'r.', )

line_i,=ax.plot(time,i,'b.')


hosp=[x*0.15 for x in i]
HOSP=[x*N for x in hosp]

line_hosp,=ax.plot(time,hosp,'k.')

line_r,=ax.plot(time,r,'g.')

ax.set_title(f'COMA Infection over time')
ax.set_ylabel('Population percentage')
ax.set_xlabel('Days')
ax.set_xlim(0,days)
#ax.set_ylim(-5,1050)
ax.xaxis.grid()
ax.yaxis.grid()

line_s.set_label('Total Susceptible')

line_i.set_label('Total Infected')

line_hosp.set_label('Required Hospitalizations')

line_r.set_label('Total Recovered')

leg=ax.legend(fontsize=16)



max_val = max(HOSP)
max_idx = HOSP.index(max_val)/24
print(f'peak hospitalization sickness occured on day {np.floor(max_idx)} with {np.floor(max_val)} infected\n')
ax.set_title(f'COMA Infection over time \n hospital peak on day {np.floor(max_idx):.0f} with {np.floor(max_val):.0f}\n Mitigation on day {mitigate} with {(1-m)*100:.2f}% effectiveness ')
plt.tight_layout()
x=1-m
fig.savefig(f'figures/COMA Corona - day {mitigate} mitigation with {x*100:.2f}% effectiveness day 63 off.tif',dpi=150)
```
